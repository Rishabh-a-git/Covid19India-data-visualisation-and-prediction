```python
cols_to_use = ['Date', 'Total Confirmed','Total Recovered','Total Deceased'] # or [0,1,2,3]
time_tot = pd.read_csv('case_time_series.csv', usecols= cols_to_use)
```

```python
time_tot.head()
```

```python
time_tot['Total Active'] = time_tot['Total Confirmed'] - (time_tot['Total Recovered'] + time_tot['Total Deceased'])
```

```python
time_tot.head()
```

```python
print(plt.rcParams.get('figure.figsize'))
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
plt.plot(time_tot['Date'], time_tot['Total Confirmed'],label = 'Confirmed',color='blue')
plt.plot(time_tot['Total Recovered'], label = 'Recovered', color='g')
plt.plot( time_tot['Total Deceased'], label = 'Deceased',color='grey')
plt.plot( time_tot['Total Active'], label = 'Active',color='red')

plt.legend()

#plt.grid(color='grey', linestyle='-', linewidth=1/4)   #Remove '#' to enable grid

plt.title('Date Wise(Total cases)')
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Cases")
plt.show()
```
