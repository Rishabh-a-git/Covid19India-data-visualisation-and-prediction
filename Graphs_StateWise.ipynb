```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
state = pd.read_csv("state_wise.csv")
```

```python
state.head()
```

```python
stno1 = state.drop(state.index[0])
```

```python
stno1a = stno1.sort_values('State')
```

```python
plt.plot(stno1['State'], stno1['Confirmed'],label = 'Confirmed',color='blue')
plt.plot( stno1['Recovered'], label = 'Recovered', color='g')
plt.plot( stno1['Active'], label = 'Active',color='red')
plt.plot( stno1['Deaths'], label = 'Deceased',color='grey')

plt.legend()

plt.title('Total Cases (till 21 May)')
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Cases")
plt.show()
```

```python
sns.set_style("ticks")
plt.figure(figsize = (15,20))
plt.barh(stno1["State"],    stno1["Confirmed"].map(int),align = 'center', color = 'lightblue', edgecolor = 'blue')
plt.xlabel('No. of cases', fontsize = 18)
plt.ylabel('States/UT', fontsize = 18)
plt.gca().invert_yaxis()
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.title('Total Confirmed Cases Statewise', fontsize = 18 )
for index, value in enumerate(stno1["Confirmed"]):
    plt.text(value, index, str(value), fontsize = 12)
plt.show()
```

```python
sns.set_style("ticks")
plt.figure(figsize = (15,20))
plt.barh(stno1["State"],    stno1["Deaths"].map(int),align = 'center', color = 'lightgrey', edgecolor = 'lightgrey')
plt.xlabel('No. of cases', fontsize = 18)
plt.ylabel('States/UT', fontsize = 18)
plt.gca().invert_yaxis()
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.title('Total Deaths Statewise', fontsize = 18 )
for index, value in enumerate(stno1["Deaths"]):
    plt.text(value, index, str(value), fontsize = 12)
plt.show()
```

```python
sns.set_style("ticks")
plt.figure(figsize = (15,20))
plt.barh(stno1["State"],    stno1["Active"].map(int),align = 'center', color = 'red', edgecolor = 'black')
plt.xlabel('No. of cases', fontsize = 18)
plt.ylabel('States/UT', fontsize = 18)
plt.gca().invert_yaxis()
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.title('Total Active Cases Statewise', fontsize = 18 )
for index, value in enumerate(stno1["Active"]):
    plt.text(value, index, str(value), fontsize = 12)
plt.show()
```

```python
sns.set_style("ticks")
plt.figure(figsize = (8,12))
plt.barh(stno1["State"],    stno1["Recovered"].map(int),align = 'center', color = 'lightgreen', edgecolor = 'green')
plt.xlabel('No. of cases', fontsize = 18)
plt.ylabel('States/UT', fontsize = 18)
plt.gca().invert_yaxis()
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.title('Total Recovered Cases Statewise', fontsize = 18 )
for index, value in enumerate(stno1["Recovered"]):
    plt.text(value, index, str(value), fontsize = 10)
plt.show()
```
